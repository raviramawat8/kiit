# Strings

Strings are combination of characters(a-z,!@$,0-9,A-Z) which can be represented by ' ' single quotes, " " double quotes, """ """ tripple quotes

```python
s1 = 'Ravi Ramawat'
s2 = "Ravi Ramawat"
s3 = """Ravi Ramawat"""
```

```python
print(s1,type(s1))
print(s2,type(s2))
print(s3,type(s3))
```
